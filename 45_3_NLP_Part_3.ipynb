- How to remove punctuation - !"#&$@<>+=/.,`[]{}()~

- How to remove Stopwords - "a","the","while" some example of stop words

```python
var = "My name is Dipak and i live in Pune$"
```

```python
import string
```

```python
print(dir(string))
```

```python
print(string.punctuation)
```

```python
punch = set(string.punctuation)
print(punch)
```

```python
Var = "My name is Dipak and i live in Pune$"
```

```python
for i in punch:
  if i in var:
    var  = var.replace(i,"")
```

```python
print(var)
```

## STOPWORDS CLEANING
- words which do not contain any useful information
- Hi I am and my name is Dipak
- Dipak
- English how many words are there
- NLTK - module in python which have a predefine Stopwords list

```python
words = ["a","an","we"]
# we are taking 3 words as stopwords
sent = "an apple fell on person head we"
```

```python
for j in words:
  if j in sent:
    sent = sent.replace(j,"")
```

```python
print(sent)
```

```python
sent = sent.split(" ")
print(sent)
```

```python
new_sent = ""
for i in sent:
  if i not in words:
    new_sent = new_sent+" "+i
```

```python
print(new_sent)
```
